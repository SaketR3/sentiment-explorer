```python
path = os.path.join(DATADIR, 'training_twitter_data.csv')
data = pd.read_csv(path, encoding='cp1252')
data
```

```python
tr_data = data.copy()
```

```python
tr_data.columns = ['col0', 'col1', 'col2', 'col3', 'col4', 'col5']
tr_data
```

```python
tr_data = tr_data.drop(columns=['col2', 'col3', 'col4'])
tr_data
```

```python
tr_data.columns = ['Sentiment', 'ID', 'Text']
tr_data
```

```python
X_train_init = tr_data.drop(columns=['Sentiment', 'ID'])
X_train_init
```

```python
binencoder = LabelBinarizer()
y_train_init = tr_data['Sentiment']
y_train_init = y_train_init.astype(int)
y_train_init = binencoder.fit_transform(y_train_init)
y_train_init
```

```python
X_train_init.isnull().sum()
```

```python
def clean(text):
    text = str(text).lower()
    text = re.sub(r'https?:\/\/\S+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

X_train_init['Text'] = X_train_init['Text'].apply(clean)
X_train_init
```

```python
X_train_full, X_test, y_train_full, y_test = train_test_split(X_train_init,
                                                    y_train_init,
                                                    test_size=0.1,
                                                    shuffle=True,
                                                    random_state=42)
```

```python
X_train_full
X_test
y_train_full
y_test
```

```python
X_train_df, y_train = X_train_full[:-50000], y_train_full[:-50000]
X_valid_df, y_valid = X_train_full[-50000:], y_train_full[-50000:]
```

```python
X_train_df
X_valid_df
y_train
y_valid
```

```python
num_y_train_zeros = 0
num_y_train_ones = 0

for i in range(len(y_train)):
  if y_train[i][0] == 0:
    num_y_train_zeros += 1
  else:
    num_y_train_ones += 1

print(num_y_train_zeros)
print(num_y_train_ones)

num_y_valid_zeros = 0
num_y_valid_ones = 0

for i in range(len(y_valid)):
  if y_valid[i][0] == 0:
    num_y_valid_zeros += 1
  else:
    num_y_valid_ones += 1

print(num_y_valid_zeros)
print(num_y_valid_ones)
```

```python
X_train = tf.data.Dataset.from_tensor_slices(X_train_df.to_dict(orient="list"))
X_valid = tf.data.Dataset.from_tensor_slices(X_valid_df.to_dict(orient="list"))
```

```python
vocab_size = 10000
text_vec_layer = tf.keras.layers.TextVectorization(max_tokens=vocab_size)
text_vec_layer.adapt(X_train.map(lambda x: x['Text']))
```

```python
text_vec_model_2 = tf.keras.Sequential([
    tf.keras.Input(shape=(1,), dtype=tf.string),
    text_vec_layer
])
```

```python
tf.random.set_seed(42)
```

```python
embed_size = 2048

model = tf.keras.Sequential([
    text_vec_layer,
    tf.keras.layers.Embedding(vocab_size, embed_size, mask_zero=True),
    tf.keras.layers.GRU(256),
    tf.keras.layers.Dense(1, activation="sigmoid")
])
```

```python
lr_schedule = tf.keras.optimizers.schedules.InverseTimeDecay(
    initial_learning_rate=0.01,
    decay_steps=10_000,
    decay_rate=1.0,
    staircase=False
)
```

```python
lr_schedule_optimizer = tf.keras.optimizers.SGD(
    learning_rate=lr_schedule,
    momentum=0.9,
    nesterov=True
)

model.compile(
    loss="binary_crossentropy",
    optimizer=lr_schedule_optimizer,
    metrics=["accuracy"]
)
```

```python
checkpoint_cb = tf.keras.callbacks.ModelCheckpoint("my_checkpoints.weights.h5", save_weights_only=True)
early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
```

```python
# train_dataset = tf.data.Dataset.from_tensor_slices((X_train_df.to_dict(orient="list"), y_train))

train_dataset = tf.data.Dataset.from_tensor_slices((X_train_df['Text'].values, y_train))
```

```python
# valid_dataset = tf.data.Dataset.from_tensor_slices((X_valid_df.to_dict(orient="list"), y_valid))

valid_dataset = tf.data.Dataset.from_tensor_slices((X_valid_df['Text'].values, y_valid))
```

```python
print("Training dataset:")
for element in train_dataset:
    print(element)
    break  # Print only the first element

print("\nValidation dataset:")
for element in valid_dataset:
    print(element)
    break  # Print only the first element
```

```python
history = model.fit(
    train_dataset.batch(32).prefetch(1),
    validation_data=valid_dataset.batch(32).prefetch(1),
    epochs=5,
    callbacks=[checkpoint_cb, early_stopping_cb]
)
```
